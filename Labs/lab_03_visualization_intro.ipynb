## Getting data into Pandas

In this case we will load data from the statsmodels.org library

MTCARS is a traditional dataset used for learning.



```python
mtcars = sm.datasets.get_rdataset("mtcars", "datasets", cache=True).data
df = pd.DataFrame(mtcars)
df['carmodel'] = df.index
print(df)
```

```python
# Calculate the count of each transmission type
transmission_counts = df['am'].value_counts()

# Create labels for the pie chart
# Assuming 'am' column has 0 for automatic and 1 for manual
transmission_labels = {0: 'Automatic', 1: 'Manual'}
labels = [transmission_labels[x] for x in transmission_counts.index]

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(transmission_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Distribution of Cars by Transmission Type')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import matplotlib.pyplot as plt

# Ensure 'df' is available, assuming it was loaded in previous cells
# If 'df' is not defined, run the data loading cells first or include them here.

# Get unique car models and create a color map for distinct colors
unique_carmodels = df['carmodel'].unique()
num_cars = len(unique_carmodels)

# Use a colormap suitable for categorical data (e.g., 'tab20' which provides 20 distinct colors, it will cycle if there are more than 20 unique cars)
colors = plt.colormaps.get_cmap('tab20') # Removed 'num_cars' as it's not a valid argument here

# Create a mapping from carmodel to a color index
color_map_indices = {model: i for i, model in enumerate(unique_carmodels)}
bar_colors = [colors(color_map_indices[model]) for model in df['carmodel']]

plt.figure(figsize=(15, 7)) # Adjust figure size for better readability
plt.bar(df['carmodel'], df['hp'], color=bar_colors)
plt.xlabel('Car Model')
plt.ylabel('Gross Horsepower (hp)')
plt.title('Gross Horsepower by Car Model')
plt.xticks(rotation=45) # Rotate x-axis labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot of wt (weight) vs. mpg
plt.figure(figsize=(8, 6))
plt.scatter(df['wt'], df['mpg'])
plt.xlabel('Weight (1000 lbs)')
plt.ylabel('Miles/(US) gallon')
plt.title('Scatter Plot of Weight vs. MPG')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate the average mpg for each cylinder count
avg_mpg_by_cyl = df.groupby('cyl')['mpg'].mean()

# Create the bar chart
plt.figure(figsize=(8, 6))
avg_mpg_by_cyl.plot(kind='bar', color='skyblue')
plt.xlabel('Number of Cylinders')
plt.ylabel('Average Miles Per Gallon (MPG)')
plt.title('Average MPG by Number of Cylinders')
plt.xticks(rotation=0) # Keep x-axis labels horizontal
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot a histogram of the weight column
plt.figure(figsize=(8, 6))
plt.hist(df['wt'], bins=10, color='lightgreen', edgecolor='black')
plt.xlabel('Weight (1000 lbs)')
plt.ylabel('Frequency')
plt.title('Distribution of Car Weight')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
